```python
plt.scatter(x, y)

plt.title("A  vs B scatter plot")
plt.xlabel("X axis ->")
plt.ylabel("Y axis ->")

text_style=dict(facecolor='red', alpha=0.5)
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 10,
        }

for pt_1, pt_2 in zip(x, y):
    plt.text(pt_1+0.2, pt_2, f"({pt_1}, {pt_2}) coordinate",# fontsize=7,
             bbox=text_style, fontdict= font)

plt.grid(axis='both',color="b", alpha=0.5, linestyle=":",
        linewidth=2)

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x, y)


font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'bold',
        'style':'italic',
        'size': 20,
        }

plt.title("A  vs B scatter plot", fontdict=font)
plt.xlabel("X axis ->", fontdict=font)
plt.ylabel("Y axis ->", fontdict=font)
# plt.xlabel("X axis ->", fontsize=20)
# plt.ylabel("Y axis ->", fontsize=20)

plt.xlim(-5,10)
plt.ylim(-10,20)

# plt.axes('off')
plt.xticks()

# plt.savefig("test.png", bbox_inches='tight')
plt.show()
```
